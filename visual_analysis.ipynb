```python
import matplotlib.pyplot as plt
import pandas as pd

```

Lets answer first question
## Which majors get the highest pay right after graduation?
This will print the top 5 carrers

```python
df = pd.read_csv('sdata.csv')
# print(df.head()) 

top_5 = df.sort_values('Starting Median Salary', ascending=False).head(5)
plt.figure(figsize=(10, 6))
plt.barh(top_5['Undergraduate Major'], top_5['Starting Median Salary'], color='green')
plt.xlabel('Starting Median Salary ($)')
plt.title('Top 5 Majors by Highest Starting Median Salary')
plt.gca().invert_yaxis()  # Highest at the top
plt.tight_layout()
plt.show()
```

## Q2 Which majors have the lowest earnings after college?

```python
top5 = df.sort_values('Starting Median Salary', ascending=True).head(5)
plt.figure(figsize=(10, 6))
plt.barh(top5['Undergraduate Major'], top5['Starting Median Salary'], color='red')
plt.xlabel('Starting Median Salary ($)')
plt.title('Top 5 Majors by Lowest Starting Median Salary')
# plt.gca().invert_yaxis()  # Highest at the top
plt.tight_layout()
plt.show()
```
