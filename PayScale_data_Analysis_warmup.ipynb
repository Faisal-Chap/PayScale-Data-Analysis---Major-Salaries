Import pandas and read the csv 'sdata' file

```python
import pandas as pd

df = pd.read_csv('/sdata.csv')


```

df.head() prints the header 5 rows

```python
print(df.head())
```

Lets get the last rows of the data to find if there is any junk data using tail()

```python
print(df.tail())
```

Now find the empty values in this data using the df.isna() it will show the true and false for every found and empty value in each cell respectively

```python
print(df.isna())
```

Now lets find the empty value as the whole in the dataframe using the df.isna().sum() that will show the comulative sum of the missing cells in each column
 - this should show the 1 for last 5 columns except the first one due to juck data at the last row

```python
print(df.isna().sum())
```

Lets clean the data using the df.drop(index=50,inplace=True)
this will will drop the last 51 row as 50+1 as the last row index and inplace means to modify the orignal data file too
- we will run this without inplace=True as this will remove the sample data and then storing it removed juck data to other variable and checking the missing value for it

```python
cleaned_df = df.drop(index=50)
print(cleaned_df.isna().sum())

# following code will remove the junk 51 row from the data
# df.drop(index=50,inplace=True)
```
