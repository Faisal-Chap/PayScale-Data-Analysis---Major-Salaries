# Pay Scale data Analysis for the University Majors to Analyze their Salary

Lets answer first question
## Which majors get the highest pay right after graduation?
This will print the top 5 carrers

```python
import pandas as pd

df = pd.read_csv('/content/sdata.csv')

soreted = df.sort_values('Starting Median Salary',ascending=False)
print(soreted.head())
```

## Q2 Which majors have the lowest earnings after college?

```python
low_earnings = df.sort_values('Starting Median Salary', ascending=True)
print(low_earnings.head())
```

## Q3 Which degrees have the highest earning potential?
Means the degree that will have the highest 90th percentile means the 90 people earn that much

```python
potential  = df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
print(potential.head())
```

## Q4 What are the lowest risk college majors from an earnings standpoint?

```python
potential  = df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
print(potential.head())
```

## Q5 Do business, STEM (Science, Technology, Engineering, Mathematics) or HASS (Humanities, Arts, Social Science) degrees earn more on average?




```python
sorted = df.groupby('Group')
print('Total Group Detail\n')
print(sorted.count()['Starting Median Salary'])

sorted_means = df.groupby('Group').mean(numeric_only=True).sort_values('Mid-Career 90th Percentile Salary', ascending=False)
print('\n\nGroup Mean\n')
print(sorted_means)
```
